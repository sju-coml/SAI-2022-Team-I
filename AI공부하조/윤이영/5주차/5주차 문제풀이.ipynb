```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import io
data = pd.read_csv(io.StringIO(uploaded['CustomerDataSet.csv'].decode('utf-8')))

print(data.head(10))
print()
print(data.info())

import numpy as np

from sklearn.preprocessing import MinMaxScaler

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#전처리
dataset = data[['ItemsBought','ItemsReturned']]

minmax_scaler = MinMaxScaler()

scaled_data= minmax_scaler.fit_transform(dataset)

print(scaled_data)

#시각화
plt.figure(figsize=(8,10))

for i in range(1,7):
  km = KMeans(n_clusters = i)
  cluster = km.fit_predict(scaled_data)
  plt.subplot(3, 2, i)
  plt.tight_layout()
  plt.title(f"K value = {i}")
  plt.scatter(scaled_data[:,0], scaled_data[:,1], c=cluster)
  plt.xlabel('ItemsBought')
  plt.ylabel('ItemsReturned')

plt.show()
```
