#문제3

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris = load_iris()

k = iris['data']

plt.figure(figsize=(5,10))
plt.subplot(211)
plt.scatter(k[:,0], k[:,1], c = 'r')
plt.xlabel('sepal length(cm)')
plt.ylabel('sepal width (cm)')
plt.subplot(212)
plt.scatter(k[:,2], k[:,3], c = 'b')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```

#문제4

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()


k = iris['data']

mean = np.mean(k, axis = 0)
std = np.std(k, axis = 0)

p = (k - mean) / std
plt.figure(figsize=(5,10))
plt.subplot(211)
plt.scatter(p[:,0], p[:,1], c = 'r')
plt.xlabel('sepal length(cm)')
plt.ylabel('sepal width (cm)')
plt.subplot(212)
plt.scatter(p[:,2], p[:,3], c = 'b')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```
