#문제3

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({'month':[ 1, 3, 5,7,9,11],'stock':[130,128,132,140,153,150]}) 
x = np.arange(0,13)
y = 2*x + 125

df['line'] = 2*df['month'] + 125

mse = np.sum((df['stock']-df['line'])**2)/len(df['stock'])

plt.scatter(df['month'],df['stock'])
plt.plot(x,y)
plt.show()

for i in range(6):
  print("%d 월 주가 = %d, 예측 주가 = %d"%(df['month'][i],df['stock'][i],df['line'][i]))

print("최종 mse : %d"%(mse))
print("12월 예상 주가 : %d"%(2*12+125))
```
